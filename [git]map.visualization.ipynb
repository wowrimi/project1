```python
# import data
df_raw = pd.read_csv("./EDA_03")

# 활용 데이터 프레임화 
df = df_raw
# del df['Unnamed: 0']
df.info()
```

```python
df2 = df[['lat','long', 'desertionno']]
```

```python
k = df2.head(10000)
k.reset_index
k
```

```python
k = k.groupby(['lat','long'])
```

```python
k = k.count()
```

```python
k.reset_index()
```

```python
k.rename(columns = {"desertionno": "count"}, inplace = True)
k
```

```python
k = k.reset_index()
```

```python
lat = k.lat.mean()
long = k.long.mean()
```

```python
import folium

#지도의 중심을 지정하기 위해 위도와 경도의 평균 구하기
lat = k['lat'].mean()
long = k['long'].mean()

from folium.plugins import HeatMap
 

m = folium.Map(
    location = [lat, long],
    zoom_start = 8,
    tiles = 'Cartodb Positron'
)
 
for i in range(0, len(k)):
    folium.Circle(
        location = [k.lat[i],k.long[i]],
        color = '#000000',
        radius = 50, 
        fill = 'crimson',
    ).add_to(m)
    
m.save('map1.html')
```
